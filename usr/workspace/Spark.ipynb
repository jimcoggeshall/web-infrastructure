```python
import pyspark
from pyspark import SparkContext, SparkConf
from pyspark.sql import SparkSession

spark = SparkSession.builder\
    .appName("test")\
    .master("local")\
    .enableHiveSupport()\
    .getOrCreate()
```

```python
df = spark.read.json("data/*.jsonl")
```

```python
import pyspark.sql.functions as func

pdf = df.filter(func.col("layers.dns.text_dns_qry_name").isNotNull())\
    .select(
        [
            "layers.dns.text_dns_qry_name",
            "layers.ip.ip_ip_dst_host"
        ]
    ).toPandas()
```

```python
pdf = df.filter(func.col("layers.dns.text_dns_qry_name").isNotNull())\
    .select(
        [
            "layers.dns.text_dns_qry_name",
            "layers.ip.ip_ip_dst_host"
        ]
    ).toPandas()
```

```python
dns = df.filter(func.col("layers.dns.text_dns_qry_name").isNotNull())
dns.groupBy("layers.ip.ip_ip_dst_host")\
    .agg(func.countDistinct("layers.dns.text_dns_qry_name").alias("qry_name_count"))\
    .sort(func.desc("qry_name_count"))\
    .toPandas()
```
